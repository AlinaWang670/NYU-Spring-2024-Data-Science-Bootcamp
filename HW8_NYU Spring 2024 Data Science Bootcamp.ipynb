### 1)Try different thresholds for computing predictions using 'Al' column.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

glass = pd.read_csv('glass.csv')
glass.head()
```

```python
glass.Type.value_counts().sort_index()
```

```python
glass.describe()
```

```python
glass.isnull().sum()
```

```python
glass['household'] = glass.Type.map({1:0, 2:0, 3:0, 5:1, 6:1, 7:1})
glass.household.value_counts()
```

```python
glass
```

```python
glass.sort_values( by = 'Al', inplace=True)
X = np.array(glass.Al).reshape(-1,1)
y = glass.household
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X,y)
```

```python
probabilities = logreg.predict_proba(X)
probabilities
```

```python
# Store the predicted probabilities of class 1.
glass['household_pred_prob'] = probabilities[:, 1]
```

```python
# Try different thresholds
from sklearn.metrics import accuracy_score, precision_score, recall_score
thresholds = [0.3, 0.4, 0.5, 0.6, 0.7]
for threshold in thresholds:
    # Predict using custom threshold
    y_pred = (probabilities[:, 1] > threshold).astype(int)
    
    # Calculate metrics
    accuracy = accuracy_score(y_true=y, y_pred=y_pred)
    precision = precision_score(y_true=y, y_pred=y_pred)
    recall = recall_score(y,pred)
    
    # Print results
    print(f"Threshold: {threshold}, Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}")
```
